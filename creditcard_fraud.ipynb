```python
import numpy as np
import seaborn as sbn
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
creditcard_df = pd.read_csv("creditcard.csv")
creditcard_df.head()
```

```python
creditcard_df.info()
```

```python
# check missing data
creditcard_df.isnull().sum()
```

```python
creditcard_df.describe()
```

```python
creditcard_df.duplicated()
```

```python
#Distribution of legit transactions and fraud transactions  0:legit - 1:fraud
creditcard_df["Class"].value_counts()
```

That numbers tells us the dataset is unbalanced

```python
# Seperate the data for analysis

legit = creditcard_df[creditcard_df.Class == 0]
fraud = creditcard_df[creditcard_df.Class == 1]

print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
creditcard_df.groupby("Class").mean()
```

We've seen that the dataset is unbalanced. it contains 284315 legit transactions and 492 fraud transactions. that is why we are going to apply undersampling method.

Built a sample dataset containing similar distribution of legit and fraud transactions  


```python
legit_sample = legit.sample(492) # randomly
```

```python
# concatenate two dataframes

new_dataframe = pd.concat([legit_sample, fraud], axis=0)
new_dataframe.head()
```

```python
# end of the datasets
new_dataframe.tail()
```

```python
new_dataframe["Class"].value_counts()
```

```python
new_dataframe.groupby("Class").mean()
```

If the samples were not good enough mean values between new and old dataframe would be not similar

```python
x = new_dataframe.drop(columns="Class",axis=1)
y = new_dataframe["Class"]
```

```python
print(x.shape)
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=42)
```

Stratified sampling can be useful when dealing with imbalanced dataset like we used to ensure that the training and test datasets have the same proportion of class labels as the input dataset

```python
print(x_train.shape, x_test.shape)
```

Train Model

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
# accuracy on training data
x_train_prediction = model.predict(x_train)
training_predict_acc = accuracy_score(x_train_prediction, y_train)
print("Prediction Accuracy on Training data :", training_predict_acc)
```

```python
# accuracy on test data

y_pred = model.predict(x_test)
test_predict_acc = accuracy_score(y_pred, y_test)
print("Prediction Accuracy on test data : ",test_predict_acc)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
c_matrix = confusion_matrix(y_test,y_pred)
sbn.heatmap(c_matrix,annot=True)
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```
